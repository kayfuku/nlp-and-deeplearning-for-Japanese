```python
import re
import pickle

# Read a file. 
with open("gingatetsudono_yoru.txt", mode="r", encoding="utf-8") as f:  
    ginga_original = f.read()

print(ginga_original)
```

```python
# Delete ruby.
ginga = re.sub("《[^》]+》", "", ginga_original)
# Delete annotation. 
ginga = re.sub("［[^］]+］", "", ginga) 
# Delete "|", full-width and half-width spaces, "「", "」", and new lines. 
ginga = re.sub("[｜ 　「」\n]", "", ginga) 

# Set "。" to a separator. 
seperator = "。"  
ginga_list = ginga.split(seperator)  
ginga_list.pop() 
# Since the last string is empty, delete it.  
ginga_list = [x+seperator for x in ginga_list]  

# Save it in pickle. 
with open('ginga_list.pickle', mode='wb') as f:  
    pickle.dump(ginga_list, f)
    
with open('ginga_list.pickle', mode='rb') as f:
    ginga_list = pickle.load(f)

print(ginga_list)
```

```python
# Tokenize. 

from janome.tokenizer import Tokenizer
import pickle

t = Tokenizer()

for sentence in ginga_list:
    print(t.tokenize(sentence, wakati=True))
```

```python
# Count words. 

import collections

t = Tokenizer()

words = []
for sentence in ginga_list:
    # Put all the words in 'words'. 
    words += t.tokenize(sentence, wakati=True)   

c = collections.Counter(words)
print(c)
```
